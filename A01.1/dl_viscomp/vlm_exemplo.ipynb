```python
import base64
from openai import OpenAI
```

```python
client_openai_local = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama",  # O Ollama ignora a chave, mas a biblioteca exige uma
)

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

image_path = "../images/face1.png"
base64_image = encode_image(image_path)

response = client_openai_local.chat.completions.create(
    model="gemma3:4b",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Qual a idade e sexo da pessoa nessa imagem?"},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/png;base64,{base64_image}"
                    },
                },
            ],
        }
    ],
)

print(f"Resp.: {response.choices[0].message.content}")
```

```python
def get_resposta_vml(image_path, cliente, pergunta):
    base64_image = None
    with open(image_path, "rb") as image_file:
        base64_image = base64.b64encode(image_file.read()).decode('utf-8')

    response = cliente.chat.completions.create(
        model="gemma3:4b",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": pergunta},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{base64_image}"
                        },
                    },
                ],
            }
        ],
    )

    return response.choices[0].message.content
```

```python
client_openai_local = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama",
)
```

```python
resp = get_resposta_vml("../images/face1.png", client_openai_local, "Qual a idade e gênero da mulher nessa imagem?")
print(resp)
```
